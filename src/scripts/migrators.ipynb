# Plots related to stellar migration in GCE models

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stars = pd.read_csv('insideout_stars.csv')
stars
```

```python
fig, ax = plt.subplots()
dr_min = 0.1
mass_in_bin = stars.groupby('galr_final')['mstar'].sum()
all_migrators = stars[(stars.dr >= dr_min) | (stars.dr <= -dr_min)].copy()
migrator_fraction = all_migrators.groupby('galr_final')['mstar'].sum() / mass_in_bin
ax.plot(migrator_fraction.index + 0.05, migrator_fraction, label='All')

outward_migrators = stars[stars.dr >= dr_min].copy()
outward_fraction = outward_migrators.groupby('galr_final')['mstar'].sum() / mass_in_bin
ax.plot(outward_fraction.index + 0.05, outward_fraction, label='Outward')

inward_migrators = stars[stars.dr <= -dr_min].copy()
inward_fraction = inward_migrators.groupby('galr_final')['mstar'].sum() / mass_in_bin
ax.plot(inward_fraction.index + 0.05, inward_fraction, label='Inward')

ax.legend(title='>%s kpc migrators' % dr_min)
plt.show()
```

```python
# Get the present-day gas metallicity gradient (by zone)
# each zone is 0.1 kpc in width, starting at 0 kpc
grad_oh = pd.read_csv('insideout_gradient.csv')
grad_oh
```

```python
fig, ax = plt.subplots()
oh_err = 0.
# stars within gas disk
subset = stars[(stars['galr_final'] < 15.5) & (stars['mass'] > 0)].copy()
# present-day gas metallicity at final radius of each star
subset['[o/h]_gas'] = grad_oh.loc[subset['zone_final']].set_axis(subset.index)
# which stars are substantially higher-metallicity
outward_migrators = subset[subset['[o/h]'] > (subset['[o/h]_gas'] + oh_err)]
ax.scatter(subset['dr'], subset['[o/h]'] - subset['[o/h]_gas'], 
           c='gray', alpha=0.2, marker='.', s=1)
ax.scatter(outward_migrators['dr'], outward_migrators['[o/h]'] - outward_migrators['[o/h]_gas'],
           c='r', marker='.', s=1)
ax.axhline(oh_err, ls='--')
ax.axvline(0, ls=':')
ax.set_xlabel(r'$\Delta R_{\rm gal}$ [kpc]')
ax.set_ylabel(r'${\rm [O/H]}_{\rm star} - {\rm [O/H]}_{\rm gas}$')
plt.show()
```

```python
fig, ax = plt.subplots()
for oh_err in [0, 0.05, 0.1, 0.15, 0.2]:
    outward_migrators = subset[subset['[o/h]'] > (subset['[o/h]_gas'] + oh_err)]
    migrator_mass = outward_migrators.groupby('galr_final')['mstar'].sum()
    total_mass = subset.groupby('galr_final')['mstar'].sum()
    ax.plot(np.arange(0.05, 15.5, 0.1), migrator_mass / total_mass, label=oh_err)
# Compare to actual migrator fraction
mass_in_bin = subset.groupby('galr_final')['mstar'].sum()
for dr_min, ls in zip([1, 2], ['-', '--']):
    outward_migrators = subset[subset.dr >= dr_min].copy()
    outward_fraction = outward_migrators.groupby('galr_final')['mstar'].sum() / mass_in_bin
    ax.plot(outward_fraction.index + 0.05, outward_fraction, c='k', ls=ls, 
            label='True migrators\n(dr>%s kpc)' % dr_min)
ax.set_title('Metallicity-determined outward migrator fraction')
ax.set_xlabel('Radius [kpc]')
ax.set_ylabel('Migrator fraction')
ax.legend(title=r'Minimum $\Delta{\rm [O/H]}$')
plt.show()
```
